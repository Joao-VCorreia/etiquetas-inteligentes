```python
import pandas as pd
import numpy as np
import random as rd
```

```python
dfPacientes = pd.read_csv("medical_records.csv")

dfPacientes = dfPacientes.iloc[:600]
dfPacientes.drop(['patient_id', 'medical_conditions', 'medications', 'allergies'], axis=1, inplace=True)
dfPacientes.rename(columns={'name':'nome', 'date_of_birth':'dataNascimento', 'gender':'sexo', 'last_appointment_date': 'ultimaConsulta'}, inplace=True)
dfPacientes = dfPacientes[['nome', 'sexo', 'dataNascimento', 'ultimaConsulta']]

dfPacientes
```

```python
letrasAZ = [chr(i) for i in range(65, 91)]

medicamentos = {}

for letra in letrasAZ:
    probabilidadeNormal = 0.7

    if rd.random() < probabilidadeNormal:
        tipo = 'normal'
    else:
        tipo = 'psicotropico'
    
    nomeMedicamento = f'Medicamento{letra}'
    medicamentos[nomeMedicamento] = tipo
```

```python
def gerarMedicacoes(num_medicacoes, medicamentos):
    meds = rd.choices(list(medicamentos.keys()), k=num_medicacoes)
    horarios = [rd.choice(pd.date_range('00:00', '23:59', freq='2h').time) for _ in range(num_medicacoes)]
    return [(med, horarios[i], medicamentos[med]) for i, med in enumerate(meds)]

numMedicacoesPorPaciente = np.random.randint(1,20, size=len(dfPacientes)) #gera um array com o mesmo numero de elementos da dfPacientes
dfPacientes['medicacoes'] = [gerarMedicacoes(num, medicamentos) for num in numMedicacoesPorPaciente]

dfPacientes.to_csv('prontuarios.csv', index=False)
```
